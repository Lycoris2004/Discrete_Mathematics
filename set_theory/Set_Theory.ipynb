# Complement

As an assignment, show that $\textbf{U}^{c} = \emptyset$ <br>

###### Proof:<br>
As per definition, complement is "The subset of some set that excludes a given subset." <br>
Universal Set is the set that contains everything. Therefore, the set that excludes the universal set is nothing or null $\emptyset$.

## Given: A and B are sets, $\exists$ x such that (x $\in$ A $\land$ x $\in$ B)<br>

a. $\textbf{A}\setminus\textbf{B} = \textbf{A} \cap \textbf{B}^c$ <br>

###### Proof:<br>
Left Side: <br>
$\textbf{A}\setminus\textbf{B} = \textbf{A} - \textbf{B}$, which basically means "what is in $\textbf{A}$ that is not in $\textbf{B}$" <br>

Right Side:<br>
$\textbf{B}^c = \textbf{U} - \textbf{B}$, which basically means "everything that is not an element of $\textbf{B}$" <br>
$\textbf{A} \cap \textbf{B}^c = \textbf{A} \cap (\,\textbf{U} - \textbf{B})\,$, which basically means "what is in $\textbf{A}$ that is not in $\textbf{B}$" <br>

Since $\textbf{A}\setminus\textbf{B} =$ "what is in $\textbf{A}$ that is not in $\textbf{B}$" <br>
And $\textbf{A} \cap \textbf{B}^c =$ "what is in $\textbf{A}$ that is not in $\textbf{B}$", <br>
Then, $\textbf{A}\setminus\textbf{B} = \textbf{A} \cap \textbf{B}^c$ <br>

b. $\textbf{B} = (\,\textbf{B} \cap \textbf{A})\, \cup (\,\textbf{B} \cap \textbf{A}^c)\, $ <br>

###### Proof:<br>
Since $\textbf{A}$ and $\textbf{B}$ have at least one element in common, we can rewrite $(\,\textbf{B} \cap \textbf{A})\, \cup (\,\textbf{B} \cap \textbf{A}^c)\, $ <br>
into $\textbf{B}  \cup (\,\textbf{A} \cap\textbf{A}^c)\, $ <br>
$\textbf{A}^c = \textbf{U} - \textbf{A}$ basically means "everything that is not in $\textbf{A}$" <br>
So $(\,\textbf{A} \cap\textbf{A}^c)\,$ is equal to $\emptyset$ <br>
Then $\textbf{B}  \cup (\,\textbf{A} \cap\textbf{A}^c)\, = \textbf{B} \cup \emptyset = \textbf{B} $ <br>
Therefore, $\textbf{B} = (\,\textbf{B} \cap \textbf{A})\, \cup (\,\textbf{B} \cap \textbf{A}^c)\, $ <br>
<br>

###### Proof:<br>

d. $\textbf{A} \cup \textbf{B} = \textbf{A} \cup (\,\textbf{B} \cap \textbf{A}^c)\, $ <br>

###### Proof:<br>
Left Side: <br>
Since there are only two variables, $\textbf{A}$ and $\textbf{B}$ <br>
$\textbf{A} \cup \textbf{B}$ = $\textbf{U}$ <br>

Right Side: <br>
$\textbf{A}^c = \textbf{U} - \textbf{A}$, which basically means "everything that is not an element of $\textbf{A}$" <br>
$(\,\textbf{B} \cap \textbf{A}^c)\, = \textbf{B} \cap (\,\textbf{U} - \textbf{A})\,$, which basically means "what is in $\textbf{B}$ that is not in $\textbf{A}$" <br>
$\textbf{A} \cup (\,\textbf{B} \cap \textbf{A}^c)\,$ then basically means "the union of $\textbf{A}$ with what is in $\textbf{B}$ that is not in $\textbf{A}$" <br>
That union is essentially just the "union of $\textbf{A}$ and $\textbf{B}$"<br>
Which was mentioned earlier that, <br>
Since there are only two variables, $\textbf{A}$ and $\textbf{B}$ <br>
$\textbf{A} \cup \textbf{B}$ = $\textbf{U}$ <br>

Since left side and right side are both equal to $\textbf{U}$ <br>
Then, $\textbf{A} \cup \textbf{B} = \textbf{A} \cup (\,\textbf{B} \cap \textbf{A}^c)\, $ <br>
<br>
<br>

e. $(\,\textbf{U}_{i=1}^{\infty} A_{i}^{c})^{c} = \cap_{i=1}^{\infty}A_{i}$ <br>

###### Proof:<br>
$\textbf{U}_{i=1}^{\infty} A_{i}$ means "the union of every partition in $\textbf{Set A}$" or just "$\textbf{Set A}$", since everything in the partition of $\textbf{Set A}$ is what makes up the set.<br>
Let us first find the answer to the inside of the parenthesis because of GEMDAS rule.<br>

Making use of the definition of complement, which is "The subset of some set that excludes a given subset." we can say that:<br>
The answer of $(\,\textbf{U}_{i=1}^{\infty} A_{i}^{c})$ is "$\textbf{Null}$ $\emptyset$" because this essentially means "the complement of $\textbf{Set A}$"<br>

Making use of the answer we got from inside the parenthesis, $(\,\textbf{U}_{i=1}^{\infty} A_{i}^{c})^{c}$ means "the complement of $\textbf{Null}$ $\emptyset$" <br>
$\textbf{Null}$ $\emptyset$ is essentially the opposite of $\textbf{U}$ <br>
Therefore, $(\,\textbf{U}_{i=1}^{\infty} A_{i}^{c})^{c} = \textbf{Set A}$ = $\textbf{U}_{i=1}^{\infty} A_{i}$<br>
<br>

f. $\vert\textbf{A}\cup\textbf{B}\vert = \vert\textbf{A}\vert + \vert\textbf{B}\vert - \vert\textbf{A}\cap\textbf{B}\vert$<br>

###### Proof:<br>
$\vert\textbf{A}\cup\textbf{B}\vert$ is basically $\vert\textbf{A}\vert + \vert\textbf{B}\vert$ if the sets were disjoint, however, our given said that $\exists$ x such that (x $\in$ A $\land$ x $\in$ B), meaning $\textbf{Set A}$ and $\textbf{Set B}$ have at least one element in common, therefore $\textbf{Set A}$ and $\textbf{Set B}$ are not disjoint.<br>
Union of sets does not double count the elements in common between sets. <br>
According to Inclusion-Exclusion Principle, we have to find the cardinality of the intersection of $\textbf{Set A}$ and $\textbf{Set B}$ and subtract it from $\vert\textbf{A}\vert + \vert\textbf{B}\vert$ in order to avoid double counting.

Since we have avoided double counting on the right side, $\vert\textbf{A}\cup\textbf{B}\vert = \vert\textbf{A}\vert + \vert\textbf{B}\vert - \vert\textbf{A}\cap\textbf{B}\vert$<br>

```python
# a

from matplotlib_venn import venn3
import matplotlib.pyplot as plt

# Given data
total_people = 120
newsweek_readers = 65
times_readers = 45
fortune_readers = 42
newsweek_and_times = 20
newsweek_and_fortune = 25
times_and_fortune = 15
all_three = 8

# Calculate the number of people who read at least one magazine
at_least_one = (
    newsweek_readers +
    times_readers +
    fortune_readers -
    newsweek_and_times -
    newsweek_and_fortune -
    times_and_fortune +
    all_three
)

# Calculate the number of people who read exactly one magazine
exactly_one = (
    (newsweek_readers - newsweek_and_times - newsweek_and_fortune + all_three) +
    (times_readers - newsweek_and_times - times_and_fortune + all_three) +
    (fortune_readers - newsweek_and_fortune - times_and_fortune + all_three)
)

# Create the Venn diagram
venn = venn3(subsets=(newsweek_readers - newsweek_and_times - newsweek_and_fortune + all_three,
                      times_readers - newsweek_and_times - times_and_fortune + all_three,
                      fortune_readers - newsweek_and_fortune - times_and_fortune + all_three,
                      newsweek_and_times - all_three,
                      newsweek_and_fortune - all_three,
                      times_and_fortune - all_three,
                      all_three),
             set_labels=('Newsweek', 'Times', 'Fortune'))

# Add labels
venn.get_label_by_id('100').set_text(str(newsweek_readers - newsweek_and_times - newsweek_and_fortune + all_three))
venn.get_label_by_id('010').set_text(str(times_readers - newsweek_and_times - times_and_fortune + all_three))
venn.get_label_by_id('001').set_text(str(fortune_readers - newsweek_and_fortune - times_and_fortune + all_three))
venn.get_label_by_id('110').set_text(str(newsweek_and_times - all_three))
venn.get_label_by_id('101').set_text(str(newsweek_and_fortune - all_three))
venn.get_label_by_id('011').set_text(str(times_and_fortune - all_three))
venn.get_label_by_id('111').set_text(str(all_three))

# Display the Venn diagram
plt.title("Venn Diagram of Magazine Readers")
plt.show()

# Output results
print("Number of people who read at least one magazine:", at_least_one)
print("Number of people who read exactly one magazine:", exactly_one)
```

```python
# b

import matplotlib.pyplot as plt
from matplotlib_venn import venn3

# Set up the data
A = 15
W = 11
R = 12
AW = 5
AR = 9
WR = 4
AWR = 3
total_cars = 25

# Calculate the values
A_only = A - AR - AW + AWR
W_only = W - AW - WR + AWR
R_only = R - AR - WR + AWR
RW_not_A = WR - AWR
AR_not_W = AR - AWR
AW_not_R = AW - AWR

only_one_option = A_only + W_only + R_only
at_least_one_option = A + W + R - AW - AR - WR + AWR
none_of_the_options = total_cars - at_least_one_option

# Create the Venn diagram
venn_labels = {'100': A_only, '010': W_only, '001': R_only, '101': AR_not_W, '110': AW_not_R, '011': RW_not_A, '111': AWR}
venn = venn3(subsets=venn_labels, set_labels=('A', 'W', 'R'))

# Display the counts in the diagram
for subset in venn_labels:
    venn.get_label_by_id(subset).set_text(venn_labels[subset])

plt.title("Venn Diagram of Car Options")
plt.show()

# Print the calculated values
print("W only =", W_only)
print("A only =", A_only)
print("R only =", R_only)
print("RW =", RW_not_A)
print("AR =", AR_not_W)
print("AW =", AW_not_R)
print("AWR =", AWR)
print("Only one option =", only_one_option)
print("At least one option =", at_least_one_option)
print("None of the options =", none_of_the_options)
```
